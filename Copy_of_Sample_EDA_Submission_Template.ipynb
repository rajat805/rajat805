## 3. ***Data Wrangling***

### Data Wrangling Code

````python

# ==========================================================
# DATA WRANGLING
# ==========================================================

# Create a copy of the original dataset
df_copy = df.copy()

# Display dataset shape before cleaning
print("Dataset Shape Before Cleaning:", df_copy.shape)

# ==========================================================
# Check duplicate values
# ==========================================================

duplicates = df_copy.duplicated().sum()
print("\nNumber of duplicate rows:", duplicates)

# Remove duplicate records
df_copy.drop_duplicates(inplace=True)

print("Dataset Shape After Removing Duplicates:",
      df_copy.shape)

# ==========================================================
# Check missing values
# ==========================================================

missing_values = df_copy.isnull().sum()
missing_values = missing_values[missing_values > 0]

print("\nColumns having missing values:")
print(missing_values)

# ==========================================================
# Calculate percentage of missing values
# ==========================================================

missing_percent = (
    df_copy.isnull().sum() /
    len(df_copy)
) * 100

missing_percent = missing_percent[
    missing_percent > 0
].sort_values(ascending=False)

print("\nPercentage of missing values:")
print(missing_percent)

# ==========================================================
# Drop columns having more than 70% missing values
# ==========================================================

threshold = 70

columns_to_drop = missing_percent[
    missing_percent > threshold
].index

print("\nColumns to drop:")
print(columns_to_drop)

df_copy.drop(
    columns=columns_to_drop,
    inplace=True
)

# ==========================================================
# Fill numerical missing values using median
# ==========================================================

numerical_cols = df_copy.select_dtypes(
    include=['int64','float64']
).columns

for col in numerical_cols:
    df_copy[col].fillna(
        df_copy[col].median(),
        inplace=True
    )

# ==========================================================
# Fill categorical missing values using mode
# ==========================================================

categorical_cols = df_copy.select_dtypes(
    include='object'
).columns

for col in categorical_cols:
    if df_copy[col].isnull().sum() > 0:
        df_copy[col].fillna(
            df_copy[col].mode()[0],
            inplace=True
        )

# ==========================================================
# Final check
# ==========================================================

print("\nRemaining Missing Values:")
print(df_copy.isnull().sum().sum())

print("\nFinal Dataset Shape:")
print(df_copy.shape)
````
